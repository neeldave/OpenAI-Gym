```python
import gym
import numpy as np
from random import random

class cartPole:
    pass

env = gym.make('CartPole-v0')
dimensions = 4
swarmSize = 5
tempW = 1
wDamp = .9
c1 = 2
c2 = 2
maxIter = 100

renderBest = cartPole()
renderBest.bestFitness = -1
renderBest.bestPosition = []
particles = []

def Fitness(p):
    length = []
    for y in range(100):
        observation = env.reset()
        done = False
        cnt = 0
        while not done:
            cnt += 1
            action = 1 if np.dot(observation, p) > 0 else 0
            observation, reward, done, info = env.step(action)
            if done:
                break
        length.append(cnt)
    averageLength = np.average(length)
    return averageLength

def ParticleSwarmOptimization():
    w = tempW
    part = cartPole()
    for i in range(swarmSize):
        part.position = np.random.uniform(0, 1, dimensions)
        part.velocity = np.random.uniform(0, 1, dimensions)
        part.fitness = Fitness(part.position)
        part.bestPosition = part.position
        part.bestFitness = part.fitness
        if part.bestFitness > renderBest.bestFitness:
            renderBest.bestFitness = part.bestFitness
            renderBest.bestPosition = part.bestPosition.copy()
        particles.append(part)

    for i in range(maxIter):
        for pp in particles:
            pp.velocity = w * pp.velocity + \
                          c1 * random() * (pp.bestPosition - pp.position) + \
                          c2 * random() * (renderBest.bestPosition - pp.position)
            x = pp.position + pp.velocity
            for sm in range(dimensions):
                if x[sm] > 1: x[sm] = random()
                if x[sm] < 0: x[sm] = random()
            pp.position = x
            pp.fitness = Fitness(pp.position)
            if pp.fitness > pp.bestFitness:
                pp.bestPosition = pp.position
                pp.bestFitness = pp.fitness
                if pp.bestFitness > renderBest.bestFitness:
                    renderBest.bestFitness = pp.bestFitness
                    renderBest.bestPosition = pp.bestPosition
        w = w * wDamp
        print(i, renderBest.bestFitness, renderBest.bestPosition)
        if renderBest.bestFitness == 200:
            break

ParticleSwarmOptimization()
print(renderBest.bestPosition)
print (renderBest.bestFitness)

best_weights = renderBest.bestPosition
avgLength = []

for i in range(10):
    observation = env.reset()
    cnt = 0
    done = False
    while not done:
        env.render()
        cnt += 1
        action = 1 if np.dot(observation, best_weights) > 0 else 0
        observation, reward, done, info = env.step(action)
        if done:
            break
    print('With best weights game ', i, ' lasted: ', cnt, 'moves')
    avgLength.append(cnt)
    
print(' ')
print(' -------------- ')
print(' ')
print('With best weights game lasted average: ', np.average(avgLength), 'moves')
env.close()
```
